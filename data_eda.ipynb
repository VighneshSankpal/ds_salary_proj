```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./ds_salary_cleaned.csv")
```

```python
df.head(4)
```

```python
df['Job Title'].value_counts()
```

```python
job_categories = {'director':['director','expert'],
                  'maneger':['manager'],                    
                  'data scientist':['data scientist','data science'],
                  'data engineer':['data engineer','warehourse'],
                  'analyst':['analyst'],
                  'mle':['machine learning','ml','mle','NLP']}
                  
```

```python
job_seniority = { 'senior':['sr','senior','principle','lead','leader',"project", "management"],
                  'junior':['jr','junior']
                }
```

```python
title = df['Job Title'].str.lower().fillna('')

df['job_category'] = pd.Series(index=df.index,dtype='object')


for category, keywords in job_categories.items():
    pattern = '|'.join(keywords)
    mask = title.str.contains(pattern, regex=True)
    df.loc[mask & df['job_category'].isna(), 'job_category'] = category
```

```python
df['seniority'] = pd.Series(index=df.index,dtype='object')

title = df['Job Title'].str.lower().fillna('')

for level,keywords in job_seniority.items():
    pattern = '|'.join(keywords)
    mask  = title.str.contains(pattern,regex=True)
    df.loc[mask & df['seniority'].isna(),'seniority'] = level
```

```python
df['job_category'] = df['job_category'].fillna('na')
df['seniority'] = df['seniority'].fillna('na')
```

```python
df.columns
```

```python
# job description len
df['Job Description'] = df['Job Description'].fillna('')
df['desc_len'] = df['Job Description'].str.len()
```

```python
# compititers counts
df['Competitors'] = df['Competitors'].replace('-1',np.nan)
df['competitors_counts'] = df.Competitors.str.split(',').str.len()
```

```python
# imputing the null value of competitor counts as -1
df['competitors_counts'] = df['competitors_counts'].fillna(-1)
df['competitors_counts'].value_counts()
```

```python
# hourly wage to annual.
 
# df['Min_salary'] = df.loc[df['Hourly']==1,'Min_salary']*2
df['Min_salary'] = np.where(df['Hourly']==1,df['Min_salary']*2,df['Min_salary'])
# df['Max_salary'] =  df.loc[df['Hourly']==1,'Max_salary']*2
df['Max_salary'] = np.where( df['Hourly']==1,df['Max_salary']*2,df['Max_salary'])
# df['Avg_salary'] = df.loc[df['Hourly']==1,['Min_salary','Max_salary']].mean(axis=1)
df['Avg_salary'] = np.where( df['Hourly']==1,df[['Min_salary','Max_salary']].mean(axis=1),df['Avg_salary'])
```
