```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./ds_salary_cleaned.csv")
```

```python
df.head(4)
```

```python
df['Job Title'].value_counts()
```

```python
job_categories = {'director':['director','expert'],
                  'maneger':['manager'],                    
                  'data scientist':['data scientist','data science'],
                  'data engineer':['data engineer','warehourse'],
                  'analyst':['analyst'],
                  'mle':['machine learning','ml','mle','NLP']}
                  
```

```python
job_seniority = { 'senior':['sr','senior','principle','lead','leader',"project", "management"],
                  'junior':['jr','junior']
                }
```

```python
title = df['Job Title'].str.lower().fillna('')

df['job_category'] = pd.Series(index=df.index,dtype='object')


for category, keywords in job_categories.items():
    pattern = '|'.join(keywords)
    mask = title.str.contains(pattern, regex=True)
    df.loc[mask & df['job_category'].isna(), 'job_category'] = category
```

```python
df['seniority'] = pd.Series(index=df.index,dtype='object')

title = df['Job Title'].str.lower().fillna('')

for level,keywords in job_seniority.items():
    pattern = '|'.join(keywords)
    mask  = title.str.contains(pattern,regex=True)
    df.loc[mask & df['seniority'].isna(),'seniority'] = level
```
