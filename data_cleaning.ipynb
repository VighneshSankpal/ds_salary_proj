### Cleaning the salary feature

```python
salary = df['Salary Estimate'].apply(lambda x: x.split('(')[0])
```

```python
minus_Kd = salary.str.replace('$','').str.replace('K','')
```

```python
minus_Kd
```

```python
df['Hourly'] = minus_Kd.apply(lambda x : 1 if 'per hour' in x.lower() else 0)
```

```python
df[df['Hourly'] ==1]
```

```python
df['salary']=minus_Kd
```

```python
df['employe provided'] =  df['salary'].apply(lambda x : 1 if 'employer provided salary' in x.lower() else 0) 
```

```python
df['salary_low_high'] = df['salary'].apply(lambda x: x.lower().replace('employer provided salary:','').replace('per hour',''))
```

```python
df[['Min_salary','Max_salary']] = df['salary_low_high'].str.split('-',expand = True).apply(pd.to_numeric
```

```python
df.drop(['salary','salary_low_high','Salary Estimate'],axis =1,inplace=True)
```
