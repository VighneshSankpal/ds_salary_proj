### Cleaning the salary feature

```python
salary = df['Salary Estimate'].apply(lambda x: x.split('(')[0])
```

```python
minus_Kd = salary.str.replace('$','').str.replace('K','')
```

```python
minus_Kd
```

```python
df['Hourly'] = minus_Kd.apply(lambda x : 1 if 'per hour' in x.lower() else 0)
```

```python
df[df['Hourly'] ==1]
```

```python
df['salary']=minus_Kd
```

```python
df['employe provided'] =  df['salary'].apply(lambda x : 1 if 'employer provided salary' in x.lower() else 0) 
```

```python
df['salary_low_high'] = df['salary'].apply(lambda x: x.lower().replace('employer provided salary:','').replace('per hour',''))
```

```python
df[['Min_salary','Max_salary']] = df['salary_low_high'].str.split('-',expand = True).apply(pd.to_numeric,errors='coerce')
```

```python
df.drop(['salary','salary_low_high','Salary Estimate'],axis =1,inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
# Create Average salary feature
df['Avg_salary'] = (df['Min_salary']/2 + df['Max_salary']/2)
```

### State feature create

```python
df['Job_state'] = df['Location'].str.split(',').str[1].str.strip().str.upper()
```

```python
df['Job_state'] = df['Job_state'].apply(lambda x : 'LA' if x == 'LOS ANGELES' else x)
```

```python
df['same_state']=(df['Headquarters'] == df['Location']).astype(int)
```

```python
#calculate age
df['company_age'] = df['Founded'].apply(lambda x : x if x<1 else 2025-x)
```

### working with job description feature

```python
def skills_present(df,col_name,skills):
    text = df[col_name].fillna('').str.lower()

    return pd.DataFrame(
        {
            f'skill_{skill}' : text.str.contains(rf'\b{skill}\b').astype(int)
            for skill in skills
        }, index = df.index
    )
    
# def skills_present(df, col_name, skills):
#     text = df[col_name].fillna('').str.lower()

#     return pd.DataFrame(
#         {
#             f'skill_{skill}': text.str.contains(rf'\b{skill}\b').astype(int)
#             for skill in skills
#         },
#         index=df.index
#     )
```

```python
job_skills=  skills_present(df,'Job Description',['python','r studio','excel','machine learning','analysis','spark','aws'])
```

```python
df_skills= pd.concat([df,job_skills],axis=1)
```

```python
df_skills['company_age']= np.where( df_skills['Founded'] <1,df_skills['Founded'],2020-df_skills['Founded'])
```

```python
df_skills.to_csv('ds_salary_cleaned.csv',index=False)
```
